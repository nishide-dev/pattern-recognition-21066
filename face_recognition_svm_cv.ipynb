```python
from PIL import Image
import numpy as np
def cross_validate(idx):
    npeople = 99  # データセットに含まれる人数の定義
    nimage = 10   #各人の画像枚数の定義
    
    imX = 92  # データベース画像の横画素数
    imY = 112 # データベース画像の縦画素数

    X  = np.zeros((0,imY*imX))
    Xt = np.zeros((0,imY*imX))
    Y  = np.array([])
    Yt = np.array([])

    X_list = X.tolist()

    for n in range(npeople):    # 人数分の顔画像についてのループ
        for i in range(nimage): # 各人の顔画像についてのループ
            if n == 97:
                img = Image.open('./self_face/' + str(n + 1) + '/' + str(i + 1) + '.png')
                img = img.resize((imX, imY))
                data = np.array(img.convert('L'))

            elif n > 92:
                img = Image.open('./self_face/' + str(n + 1) + '/' + str(i + 1) + '.jpg')
                img = img.resize((imX, imY))
                data = np.array(img.convert('L'))

            else:
                img = Image.open('./face/' + str(n + 1) + '/' + str(i + 1) + '.png')
                img = img.resize((imX, imY))
                data = np.array(img)                   # 画像をnumpy形式に変換する

            # mean, std = np.mean(data), np.std(data)
            # data = (data - mean) / std
            data = data.reshape([-1, imY*imX])

            if (i == idx):                           #  一枚目の画像についての処理(テストデータ)
                Xt = np.append(Xt, data, axis=0)   #行ベクトルを連結する
                Yt = np.hstack((Yt, np.array([n + 1])))        #行ベクトルを連結する
            else:                                  #  二枚目の画像についての処理（トレーニングデータ）
                X_list.append(data)     #行ベクトルを連結する
                Y = np.hstack((Y, np.array([n + 1])))          #行ベクトルを連結する

    X = np.asarray(X_list).reshape(-1,imY*imX)

    # scikitlearnの必要なライブラリを読み込む
    from sklearn.svm import SVC

    # SVC のインスタンスを生成する
    # svm = SVC(random_state=1, gamma='scale', kernel='rbf', C=10.0)
    svm = SVC(random_state=1, gamma='auto')

    # svm のトレーニングを行う
    svm.fit(X, Y)

    # テストデータを予測する
    Yt_pred = svm.predict(Xt)

    # 予測の成功、失敗をカウントする, 予測結果を出力する
    success = 0
    error = 0
    for test, pred in zip(Yt, Yt_pred):
        if test == pred:
            success += 1
        else:
            error += 1
    return success / (success + error)

#     print("Success: {}".format(success))
#     print("Error: {}".format(error))
#     print("Recognition accuracy = {}/{} = {}".format(success, success + error, success / (success + error)))
```

```python
import time
start = time.time()
accs = np.array([])
for i in range(10):
    acc = cross_validate(i)
    accs = np.append(accs, acc)
    print(f'{i}: acc: {acc}')
    section = time.time()
    print(f'section_time: {section - start}')
print(f'\nmean: {accs.mean()}')
```

```python
from PIL import Image
import numpy as np
def cross_validate_scaled(idx: int) -> float:
    npeople = 99  # データセットに含まれる人数の定義
    nimage = 10   #各人の画像枚数の定義
    
    imX = 92  # データベース画像の横画素数
    imY = 112 # データベース画像の縦画素数

    X  = np.zeros((0,imY*imX))
    Xt = np.zeros((0,imY*imX))
    Y  = np.array([])
    Yt = np.array([])

    X_list = X.tolist()

    for n in range(npeople):    # 人数分の顔画像についてのループ
        for i in range(nimage): # 各人の顔画像についてのループ
            if n == 97:
                img = Image.open('./self_face/' + str(n + 1) + '/' + str(i + 1) + '.png')
                img = img.resize((imX, imY))
                data = np.array(img.convert('L'))

            elif n > 92:
                img = Image.open('./self_face/' + str(n + 1) + '/' + str(i + 1) + '.jpg')
                img = img.resize((imX, imY))
                data = np.array(img.convert('L'))

            else:
                img = Image.open('./face/' + str(n + 1) + '/' + str(i + 1) + '.png')
                img = img.resize((imX, imY))
                data = np.array(img)                   # 画像をnumpy形式に変換する

            mean, std = np.mean(data), np.std(data)
            data = (data - mean) / std
            data = data.reshape([-1, imY*imX])

            if (i == idx):                           #  一枚目の画像についての処理(テストデータ)
                Xt = np.append(Xt, data, axis=0)   #行ベクトルを連結する
                Yt = np.hstack((Yt, np.array([n + 1])))        #行ベクトルを連結する
            else:                                  #  二枚目の画像についての処理（トレーニングデータ）
                X_list.append(data)     #行ベクトルを連結する
                Y = np.hstack((Y, np.array([n + 1])))          #行ベクトルを連結する

    X = np.asarray(X_list).reshape(-1,imY*imX)

    # scikitlearnの必要なライブラリを読み込む
    from sklearn.svm import SVC

    # SVC のインスタンスを生成する
    svm = SVC(random_state=1, gamma='auto', kernel='rbf')

    # svm のトレーニングを行う
    svm.fit(X, Y)

    # テストデータを予測する
    Yt_pred = svm.predict(Xt)

    # 予測の成功、失敗をカウントする, 予測結果を出力する
    success = 0
    error = 0
    for test, pred in zip(Yt, Yt_pred):
        if test == pred:
            success += 1
        else:
            error += 1
    return success / (success + error)
```

```python
start = time.time()
accs = np.array([])
for i in range(10):
    acc = cross_validate_scaled(i)
    accs = np.append(accs, acc)
    print(f'{i}: acc: {acc}')
    section = time.time()
    print(f'section_time: {section - start}')
print(f'\nmean: {accs.mean()}')
```
